```python
import os
print(os.getcwd())

# os.chdir(os.path.join(os.getcwd(), 'admire', 'models'))

from admire.models.RTDataHandler import RTDataHandler


NODE = 'e1534'

test_data = 'data/processed/turin_demo_top200'
model_config = {
    'model': 'LSTMCNN',
    'run_id': 'e_-2024_05_08-14_28_22',
    'entity': "ignacysteam",
    'project': "lightning_logs",
    'model_tag': "v0",
    'data_dir': 'data/processed/turin_demo_top200/history',
    'data_normalization': True,
    'slide_length': 1,
    'nodes_count': 200
}
```

```python
#For testing purposes feeding date from bash script
rtdatahandler = RTDataHandler(data_dir=test_data, inference_model_config=model_config, debug_printing=False, just_debug=True)
rtdatahandler.run()
```

```python
hist_ts_raw, hist_nodenames, hist_dates_range = rtdatahandler.get_history()
print(f'Historical time series shape: {hist_ts_raw.shape}')
print(f'Historical node names: {hist_nodenames}')
print(f'Historical dates range: {hist_dates_range}')
```

```python
hist_ts = rtdatahandler.transform.denormalize_time_series(hist_ts_raw)
```

```python
# data_batch = rtdatahandler.()
# print(f'New data batch shape: {data_batch.shape}')
```

```python
hist_nodename_ts_map = {node: data.T for node, data in zip(hist_nodenames, hist_ts)}
```

```python
# Get the data for the NODE
node_data = hist_nodename_ts_map[NODE]

print(f'Node data shape: {node_data.shape}')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px


def plot_single_node_data(node_data: np.ndarray, node_name: str):
    # Plot the data, four channels, each having a separate plot
    data = node_data
    print(data.shape)
    
    fig, axs = plt.subplots(4, 1, figsize=(15, 6))
    for i in range(4):
        axs[i].plot(data[:, i])
        axs[i].set_title(f'Channel {i}')
        if i == 3:
            axs[i].set_xlabel('Time')
        axs[i].set_ylabel('Value')
    plt.suptitle(f'Node {node_name} data')
    plt.tight_layout()
    plt.show()
    
def plot_all_nodes_data(hist_nodename_ts_map: dict, node_to_highlight: str, frac: float = 1.0):
    highlight_data = hist_nodename_ts_map[node_to_highlight].T
    print(highlight_data.shape)
    data = np.array(list(hist_nodename_ts_map.values()))
    indices = np.random.choice(data.shape[0], int(frac * data.shape[0]), replace=False)
    data = data[indices]
    
    fig, axs = plt.subplots(4, 1, figsize=(15, 6))
    for i in range(4):
        
        if i == 0:
            axs[i].plot(data[i], color='gray', alpha=0.5, label='Other nodes')
        else:
            axs[i].plot(data[i], color='gray', alpha=0.5)
        
        axs[i].plot(highlight_data[i], color='red', lw=2, label=node_to_highlight)
        
        axs[i].set_title(f'Channel {i}')
        if i == 3:
            axs[i].set_xlabel('Time')
        axs[i].set_ylabel('Value')
        
        # axs[i].set_ylim([0, 1])
    

    plt.legend()
    plt.suptitle('Data for all nodes')
    plt.tight_layout()
    plt.show()
    
    
```

```python
plot_single_node_data(node_data, NODE)
```

```python
plot_all_nodes_data(hist_nodename_ts_map, NODE)
```

```python
redis_data = rtdatahandler.redis_data

# plot histogram 
redis_data_names = list(redis_data.keys())
redis_data_values = list(redis_data.values())

fig = px.bar(x=redis_data_names, y=redis_data_values)
# Make the NODE data red
fig.data[0].marker.color = ['red' if name == NODE else 'gray' for name in redis_data_names]
fig.show()
```
